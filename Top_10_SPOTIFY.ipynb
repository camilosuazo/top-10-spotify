```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Subir el archivo
uploaded = files.upload()

# Cargar el archivo en un DataFrame
for filename in uploaded.keys():
    spotify_data = pd.read_csv(filename)

# Ver las primeras filas del dataset
print(spotify_data.head())
```

```python
# Mostrar nombres de columnas
print(spotify_data.columns)

# Ver información general del dataset
print(spotify_data.info())

# Ver estadísticas básicas de las columnas numéricas
print(spotify_data.describe())
```

```python
# Analizamos cuáles son las 10 canciones con más reproducciones
top_10_streams = spotify_data.sort_values(by='Streams', ascending=False).head(10)
print("\nTop 10 canciones con más streams:")
print(top_10_streams[['Artist and Title', 'Streams']])
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.barh(top_10_streams['Artist and Title'], top_10_streams['Streams'], color='blue')
plt.xlabel('Streams Totales')
plt.ylabel('Canciones')
plt.title('Top 10 Canciones con Más Streams Totales')
plt.gca().invert_yaxis()  # Invertir el eje para que la canción con más streams aparezca arriba
plt.show()
```
